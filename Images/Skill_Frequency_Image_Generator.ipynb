```python
import json
import matplotlib.pyplot as plt
from pprint import pprint 

# Load the JSON file
with open("2_top_paying_job_skills.json", "r") as f:
    data = json.load(f)

pprint(data)
print()
print(f"No. of skills: {len(data)}")
```

```python
# Extract skills and counts
skills = [item["skills"].upper() for item in data]
counts = [int(item["count_of_skills"]) for item in data]
```

```python
# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(skills, counts)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Skills")
plt.ylabel("Demand Count")
plt.title("Most In-Demand Skills for Top-Paying Data Analyst Roles")
```

```python
plt.figure(figsize=(12, 6))
plt.barh(skills[:10], counts[:10])

plt.xlabel("Skill Frequency")
plt.ylabel("Skills")
plt.title("Skill Frequency in Top-Paying Data Analyst Jobs")

plt.gca().invert_yaxis()
plt.tight_layout()

# Save the figure
plt.savefig("top_paying_data_analyst_skills.png", dpi=300)
plt.show()
```
